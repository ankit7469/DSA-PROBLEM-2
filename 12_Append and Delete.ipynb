```python
# Append and delete - 


s = input()
t = input()
k = int(input())

count = 0                                         # common prefix  
for i in range(min(len(s), len(t))):
    if s[i] == t[i]:
        count += 1
    else:
        break

delete_count = len(s) - count                     # for delete
append_count = len(t) - count                     # for append
total = delete_count + append_count

if k >= len(s) + len(t):
    print("Yes")
elif k < total:
    print("No")
elif (k - total) % 2 == 0:
    print("Yes")
else:
    print("No")

```
