```python
# Jumping on the clouds --

n, k = map(int, input().split())
c = list(map(int, input().split()))
e = 100

i = 0
while True:
    i = (i + k) % n       # jump
    e -= 1               # energy for jump -1

    if c[i] == 1:        # thundercloud, energy would be -2
        e -= 2

    if i == 0:           # back to start
        break

print(e)
```

```python
Sample Input

STDIN             Function
-----             --------
8 2               n = 8, k = 2
0 0 1 0 0 1 1 0   c = [0, 0, 1, 0, 0, 1, 1, 0]
Sample Output

92
```
