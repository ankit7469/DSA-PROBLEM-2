```python
# Angry professor ---
# If the class is cancelled print YES  otherwise NO

t=int(input())                                       # Test cases
for _ in range(t):
    n,k=map(int,input().split())                     # number of students and cancellation threshold value
    arr_time=list(map(int,input().split()))          # Arrival time of students
    count=0

    for i in arr_time:

        if( i <= 0):
            count+=1
        else:
            count+=0

    if (count >= k):
        print("YES")
    else:
        print("NO")
            
```
