```python
# Circulation Array Rotation --

# n= the number of elements in the integer array, k= the rotation count and q= the number of queries
n,k,q=map(int,input().split())                
arr=list(map(int,input().split()))
print(arr)

k=k%len(arr)
for _ in range(k):
    arr.insert(0,arr.pop())

for _ in range(q):
    idx=int(input())
    print(arr[idx])
```
