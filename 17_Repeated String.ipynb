```python
# Repeated String ---
s=input()
n=int(input())

repeat=""                               # for storing string

while len(repeat) <n:
    repeat+=s                            # store values in repeat 
    repeat=repeat[:n]
    
b=repeat.count('a')                     # count a
print(b)
```

```python
# Fast 

s = input()
n = int(input())

count_a = s.count('a')                   # for counting a 
length = len(s)                          # length of string

full = n // length                       # quotient
rem = n % length                         # remainder

total = count_a * full                   # count a * quotient it means first we count a in main string , then we multiply 
total += s[:rem].count('a')              # now in extra part in remainder, we check there is a or not

print(total)
```
