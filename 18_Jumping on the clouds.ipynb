```python
# Jumping on the clouds

n=int(input())
arr=list(map(int,input().split()))

jump=0                                                # how much step jump wants e.g.=1
i=0                                                   # current position of jump

while i<n-1:
    if i+2< n and arr[i+2]==0:
        i+=2
    else:
        i+=1
    jump+=1

              
print(jump)
```

```python
Sample Input 0

7
0 0 1 0 0 1 0
Sample Output 0

4
```
