```python
# Queen's Attack II

n, k = map(int, input().split())                          # Board and obstacle
rq, cq = map(int, input().split())                        # Queen's position

# Initial maximum moves in 8 directions (till board edge)

up = n - rq
down = rq - 1
right = n - cq
left = cq - 1

up_right = min(n - rq, n - cq)
up_left = min(n - rq, cq - 1)
down_right = min(rq - 1, n - cq)
down_left = min(rq - 1, cq - 1)

# Process each obstacle
for _ in range(k):
    ro, co = map(int, input().split())                  # obstacle position

# Same column
    if co == cq:
        if ro > rq:                                     # obstacle above
            up = min(up, ro - rq - 1)
        else:                                           # obstacle below
            down = min(down, rq - ro - 1)

# Same row
    elif ro == rq:
        if co > cq:                                     # obstacle right
            right = min(right, co - cq - 1)
        else:                                           # obstacle left
            left = min(left, cq - co - 1)

# Same diagonal
    elif abs(ro - rq) == abs(co - cq):

# up-right
        if ro > rq and co > cq:
            up_right = min(up_right, ro - rq - 1)

# up-left
        elif ro > rq and co < cq:
            up_left = min(up_left, ro - rq - 1)

# down-right
        elif ro < rq and co > cq:
            down_right = min(down_right, rq - ro - 1)

# down-left
        else:
            down_left = min(down_left, rq - ro - 1)

# Total attackable squares

total = (up + down + left + right + up_right + up_left + down_right + down_left)
print(total)
```
