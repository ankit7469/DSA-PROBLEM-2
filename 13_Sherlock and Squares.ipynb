```python
# Sherlock and Squares --

import math
n=int(input())

for _ in range(n):
    count=0
    a,b=map(int,input().split())
    for i in range(a,b+1):
        m=i**0.5
        
        if m.is_integer():
            count+=1

    print(count)

```

```python
Smaple Input
2
3 9
17 24
Sample Output

2
0
```

```python
# 2nd method fast and reliable --

import math

n = int(input())

for _ in range(n):
    a, b = map(int, input().split())
    
    start = math.ceil(math.sqrt(a))
    end = math.floor(math.sqrt(b))
    
    print(end - start + 1)
```
