```python
# Cut the Sticks --

n=int(input())
arr=list(map(int,input().split()))

while len(arr)>0:
    print(len(arr))

    a=min(arr)

    arr2=[]                                         # for subtract
    for x in arr:
        arr2.append(x-a)

    arr3=[]                                         # for removing 0
    for y in arr2:
        if y!=0:
            arr3.append(y)

    arr=arr3
        

    
```
