```python
# Sequence array rotation - 

n=int(input())
p=list(map(int,input().split()))         # permutation list
for x in range(1,n+1):
    a=p.index(x)+1                       # first p() value    p[i] == x
    y=p.index(a)+1                       # second p(y) claue  p[j] == first
    print(y)

                 

```
