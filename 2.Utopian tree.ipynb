```python
# 2. Utopian tree

n=int(input())                         # number of test cases


for _ in range(n):
    cycle=int(input())
    height=1                           # initial height
    
    for i in range(cycle):
        if(i%2==0):                     # spring (even cycle) 
            height*=2
        else:
            height+=1                   # summer (odd cycle)
    print(height)
```
