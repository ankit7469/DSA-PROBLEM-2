```python
# Beautiful days at the movies ---

i, j, k = map(int, input().split())
count = 0

for n in range(i, j + 1):
    temp = n
    rev = 0

    while temp > 0:
        digit = temp % 10
        rev = rev * 10 + digit
        temp = temp // 10

    if abs(n - rev) % k == 0:
        count += 1

print(count)
```
