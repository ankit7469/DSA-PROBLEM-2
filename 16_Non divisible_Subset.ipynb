```python
# Non divisible subset --- Brute force


n,k=map(int,input().split())
s=list(map(int,input().split()))

subset=[]
for i in s:
    valid=True

    for x in subset:                                   # false condition if remainder=0 condition false
        if ((x+i)%k==0):
            valid=False
            break

    if valid:
        subset.append(i)

print(subset)
print(len(subset))

        
```

```python
# Simple and reliable -- 

n, k = map(int, input().split())
s = list(map(int, input().split()))

freq = [0] * k

for num in s:
    freq[num % k] += 1

result = 0

if freq[0] > 0:
    result += 1

for r in range(1, k // 2 + 1):
    if r != k - r:
        result += max(freq[r], freq[k - r])
    else:
        if freq[r] > 0:
            result += 1

print(result)
```
