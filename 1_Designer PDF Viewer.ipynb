```python
# Designer PDF Viewer --     Using store every value seperately

arr = list(map(int, input().split()))
str1 = input()

values = []

for ch in str1:
    if ch == 'a':
        values.append(arr[0])
    elif ch == 'b':
        values.append(arr[1])
    elif ch == 'c':
        values.append(arr[2])
    elif ch == 'd':
        values.append(arr[3])
    elif ch == 'e':
        values.append(arr[4])
    elif ch == 'f':
        values.append(arr[5])
    elif ch == 'g':
        values.append(arr[6])
    elif ch == 'h':
        values.append(arr[7])
    elif ch == 'i':
        values.append(arr[8])
    elif ch == 'j':
        values.append(arr[9])
    elif ch == 'k':
        values.append(arr[10])
    elif ch == 'l':
        values.append(arr[11])
    elif ch == 'm':
        values.append(arr[12])
    elif ch == 'n':
        values.append(arr[13])
    elif ch == 'o':
        values.append(arr[14])
    elif ch == 'p':
        values.append(arr[15])
    elif ch == 'q':
        values.append(arr[16])
    elif ch == 'r':
        values.append(arr[17])
    elif ch == 's':
        values.append(arr[18])
    elif ch == 't':
        values.append(arr[19])
    elif ch == 'u':
        values.append(arr[20])
    elif ch == 'v':
        values.append(arr[21])
    elif ch == 'w':
        values.append(arr[22])
    elif ch == 'x':
        values.append(arr[23])
    elif ch == 'y':
        values.append(arr[24])
    elif ch == 'z':
        values.append(arr[25])

high = max(values)
count = len(str1)
mul = high * count

print(mul)
```

```python
# 2nd method- using store all the value together (Using ASCII number)

arr=list(map(int,input().split()))
s=input()


values = []
for ch in s:
    values.append(arr[ord(ch) - ord('a')])          # ( ASCII NUMBER 97 - ord('A') 97 ==0   ( B)
high=max(values)  
count=len(s)

mul=high*count
print(mul)
```
